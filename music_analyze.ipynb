Clearing dataframe

```python
analys['key'] = analys['key'].replace('', np.NaN)
analys['instrumentalness'] = analys['instrumentalness'].replace('', np.NaN)
analys.dropna(subset = ['key'], inplace = True)
analys.dropna(subset = ['instrumentalness'], inplace = True)
analys.isna().sum()
```

```python
analys['key'] = analys['key'].astype(int)
analys['instrumentalness'] = analys['instrumentalness'].astype(float)
```

```python
corr = analys.corr()
corr.style.background_gradient(cmap='coolwarm')
```

```python
analys = (analys - analys.min())/(analys.max() - analys.min())
analys['Class'] = analys['Class'] * 10
analys['Class'] = analys['Class'].astype(int)
```

```python
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
```

```python
x = []
plt_ener = []
plt_acous = []
plt_pop = []
plt_loud = []
```

```python
for i in range(0,11):
  temp_df = analys[analys['Class'] == i].sort_values(['energy'], ascending = True)
  x.append(np.asarray([i for i in range(1, len(temp_df.index) + 1)]))
  plt_ener.append(np.asarray(temp_df['energy'].tolist()))
  plt_acous.append(np.asarray(temp_df['acousticness'].tolist()))
  plt_pop.append(np.asarray(temp_df['Popularity'].tolist()))
  plt_loud.append(np.asarray(temp_df['loudness'].tolist()))
```

```python
fig, ax = plt.subplots(11,1, figsize= (20,120))
ax = ax.flatten()
for i in range(0,11):
  ax[i].plot(x[i], plt_ener[i], color = 'blue' , label = 'energy' , alpha = 0.6)
  ax[i].plot(x[i], plt_acous[i], color = 'red', label = 'acousticness', alpha = 0.6)
  ax[i].plot(x[i], plt_pop[i], color = 'y', label = 'Populatiry', alpha = 0.6)
  ax[i].plot(x[i], plt_loud[i], color = 'c', label = 'loudness', alpha = 0.6)

  ax[i].set_title(f"Class {i}", fontsize = 20)
  ax[i].set_xlabel('Track')
  ax[i].set_ylabel('F')
  ax[i].legend(loc = 'lower left')

```

```python
math_wait = {'Class' : list(i for i in range(0,11)), 'energy_mw' : [], 'acousticness_mw' : [], 'Popularity_mw' : [], 'loudnes_mw' : []}
sigma = {'Class' : list(i for i in range(0,11)),'energy_sig' : [], 'acousticness_sig' : [], 'Popularity_sig' : [], 'loudness_sig' : []}
```

```python
for i in range(0,11):
  temp_df = analys[analys['Class'] == i]
  math_wait['energy_mw'].append(temp_df['energy'].mean())
  math_wait['acousticness_mw'].append(temp_df['acousticness'].mean())
  math_wait['Popularity_mw'].append(temp_df['Popularity'].mean())
  math_wait['loudnes_mw'].append(temp_df['energy'].mean())
mw_df = pd.DataFrame(data=math_wait)
mw_df
```

```python
for i in range(0,11):
  temp_df = analys[analys['Class'] == i]
  sigma['energy_sig'].append(temp_df['energy'].var(ddof=0))
  sigma['acousticness_sig'].append(temp_df['acousticness'].var(ddof=0))
  sigma['Popularity_sig'].append(temp_df['Popularity'].var(ddof=0))
  sigma['loudness_sig'].append(temp_df['loudness'].var(ddof=0))
sig_df = pd.DataFrame(data=sigma)
sig_df
```

```python
df_merged = pd.merge(mw_df, sig_df)
df_merged
```

```python
hist = df_merged.hist(figsize = (10,10), bins=10)
```
